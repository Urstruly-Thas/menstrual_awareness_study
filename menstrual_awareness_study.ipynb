```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.info()
femella_colors =[
    "#DA70D6",  # Orchid
    "#BA55D3",  # Medium Orchid
    "#9932CC",  # Dark Orchid
    "#800080",  # Purple
    "#8A2BE2",  # Blue Violet
    "#4B0082",  # Indigo
    "#6A0DAD",  # Amethyst
    "#7B1FA2",  # Deep Purple (Material Design)
]
df.columns = df.columns.str.strip()

teal_hues = [
    "#E0F7F4",  # Light Aqua Teal
    "#B2DFDB",  # Soft Teal
    "#80CBC4",  # Light Teal
    "#4DB6AC",  # Medium Teal
    "#26A69A",  # Classic Teal
    "#009688",  # Standard Teal (Material Design)
    "#00897B",  # Deep Teal
    "#00796B",  # Strong Teal
    "#00695C",  # Dark Teal
    "#004D40",  # Rich Teal
]


pink_hues = [
    "#FFE4E1",  # Misty Rose
    "#FFB6C1",  # Light Pink
    "#FF69B4",  # Hot Pink
    "#FF1493",  # Deep Pink
    "#FF85A2",  # Soft Rose Pink
    "#FF6EB4",  # Bubblegum Pink
    "#F06292",  # Medium Pink (Material Design)
    "#EC407A",  # Bright Pink
    "#E91E63",  # Raspberry Pink
    "#C2185B",  # Dark Pink / Crimson Pink
]
```

```python
# column name
col = df['Have you ever learned about menstruation in school or elsewhere?']

# Calculate percentage frequency
value_counts = col.value_counts(normalize=True) * 100
value_counts = value_counts.sort_values(ascending=False)

# Colors
colors = femella_colors[:len(value_counts)]

# Plotting
fig, ax = plt.subplots(figsize=(8, 5))
bars = sns.barplot(x=value_counts.index, y=value_counts.values, ax=ax, palette=colors, width=0.7 )
#adding edge color to the bars
for bar in bars.patches:
    bar.set_edgecolor("#4B0076")
    bar.set_linewidth(1.5)

# Add percentage labels on top of bars
for bar in bars.patches:
    height = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        height + 1,
        f'{height:.1f}%',
        ha='center',
        va='bottom',
        fontsize=12,
        fontweight='bold',
        color='#4B0076'
    )

# Aesthetics
ax.set_title("Have you ever learned about menstruation in school or elsewhere?", fontsize=12, fontweight='bold', color='black',pad=30)
ax.set_ylabel("Percentage of respondents", fontsize=12, color ="black" )
ax.set_xlabel("")
sns.despine(left=False, bottom=False, ax=ax)
ax.spines['bottom'].set_color("black")
ax.spines['left'].set_color("black")
ax.tick_params(axis='x', labelrotation=0)
ax.tick_params(colors="black")
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0f}%'))

ax.yaxis.grid(True, color="black", linestyle='-', linewidth=0.7)
ax.set_axisbelow(True)

filename = "learned_about_menstruation.png"
plt.tight_layout()
plt.savefig(filename, dpi=300, bbox_inches='tight', transparent=True)
plt.show()

# Download in Colab
files.download(filename)
```

```python
# Extract the column with the responses
col = 'Do you feel comfortable discussing menstruation with:'
responses = df[col].dropna()

# Define the categories to count
categories = ['Friends', 'Family members', 'Partner', 'Colleagues',
              'I don’t feel comfortable', 'I’ve never thought about it']

# Count occurrences across all responses
from collections import Counter

category_counts = Counter()
for response in responses:
    for category in categories:
        if category in response:
            category_counts[category] += 1

# Merge related categories (if needed)
category_counts['I don’t feel comfortable / Never thought about it'] = \
    category_counts.pop('I don’t feel comfortable', 0) + category_counts.pop('I’ve never thought about it', 0)

# Create labels and values
labels = list(category_counts.keys())
values = list(category_counts.values())


plt.figure(figsize=(20, 5))
wedges, _, autotexts = plt.pie(
    values,
    colors=teal_hues[:len(labels)],
    autopct='%1.1f%%',
    startangle=180,
    textprops={'color': 'black'}
)

plt.legend(wedges, labels, title="Comfortable With", loc="center left", bbox_to_anchor=(0.9, 0.5), fontsize=9, title_fontsize=10, handlelength=1.2,
    handletextpad=0.2,
    borderaxespad=0.2,frameon = False )
plt.setp(autotexts, size=12, weight="bold")

# Title and formatting
plt.title("Who Do People Feel Comfortable Discussing Menstruation With?", fontsize=14, color='black')
plt.axis('equal')

# Save with transparent background
plt.savefig("menstruation_comfort_piechart.png", transparent=True)
plt.tight_layout()
plt.show()

files.download("menstruation_comfort_piechart.png")
```

```python

from collections import Counter
col = "Are there cultural/ social restrictions placed on you during your periods?"

# Drop NaNs
responses = df[col].dropna()

# Count individual items (split by comma)
category_counts = Counter()
for response in responses:
    parts = [item.strip() for item in response.split(',')]
    category_counts.update(parts)

# Convert to percentage
total_responses = len(responses)
percentages = {k: (v / total_responses) * 100 for k, v in category_counts.items()}

# Show results
print(category_counts)
for k, v in sorted(percentages.items(), key=lambda x: x[1], reverse=True):
    print(f"{k}: {v:.2f}%")
```

```python
# Set the visual style
sns.set(style="whitegrid")
plt.rcParams.update({'font.size': 13, 'font.family': 'sans-serif'})

# Data for both questions
data = {
    'Response': ['Yes', 'No', 'Maybe'],
    'Heard About Stigma (%)': [65.6, 23.7, 10.8],
    'Home/Community Restrictions (%)': [58.1, 24.7, 16.1]
}

# Create a dataframe
df = pd.DataFrame(data)

# Reshape the data for seaborn
df_melted = df.melt(id_vars='Response', var_name='Question', value_name='Percentage')

# Create the diverging bar plot
plt.figure(figsize=(10, 6))
palette = ['#BA55D3', '#4B0082']

ax = sns.barplot(
    x='Percentage',
    y='Response',
    hue='Question',
    data=df_melted,
    palette=palette
)
#Add percentage text on top of each bar
for container in ax.containers:
    for bar in container:
        height = bar.get_width()
        ax.text(
            height + 1,                        # Position text slightly to the right
            bar.get_y() + bar.get_height()/2, # Vertically center it on the bar
            f'{height:.1f}%',                 # Format as percentage
            va='center',
            ha='left',
            fontsize=11,fontweight='bold',
            color='#4B0076'
        )

# Aesthetic enhancements
plt.title("We’ve Heard the Stigma—But Have We Unlearned It?", fontsize=16, weight='bold')
plt.xlabel("Percentage of Respondents")
plt.ylabel("")
plt.xlim(0, 100)
plt.legend(title='', loc='lower right', frameon=False)
plt.grid(axis='x', linestyle='-', color = "Black")
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0f}%'))
# Make bottom and left spines black
ax.spines['bottom'].set_color('black')
ax.spines['left'].set_color('black')

# Optional: Hide top and right spines for a cleaner look
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Save the figure
plt.tight_layout()
plt.savefig("menstruation_diverging_bar.png", dpi=300, transparent=True)

# Show the chart
plt.show()
files.download("menstruation_diverging_bar.png")
```

```python

# column name
column = "Do you think menstrual leave should be legally mandated in India across all sectors?"

# Value counts and percentages
counts = df[column].value_counts(dropna=True)
percentages = (counts / counts.sum() * 100).round(2)

# Colors
femella_colors = [
    "#DA70D6",  # Orchid
    "#BA55D3",  # Medium Orchid
    "#9932CC",  # Dark Orchid
    "#800080",  # Purple
    "#8A2BE2",  # Blue Violet
    "#4B0082",  # Indigo
    "#6A0DAD",  # Amethyst
    "#7B1FA2",  # Deep Purple (Material Design)
]

# Plot
fig, ax = plt.subplots(figsize=(8, 4))
bars = ax.barh(percentages.index, percentages.values, color=femella_colors[:len(percentages)])

# Add percentage labels
for bar, percent in zip(bars, percentages):
    width = bar.get_width()
    ax.text(width + 1, bar.get_y() + bar.get_height() / 2,
            f'{percent:.1f}%', va='center', fontsize=11, color="#800080", fontweight="bold")

# Aesthetics
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('black')
ax.spines['bottom'].set_color('black')

ax.tick_params(axis='x', bottom=False, labelbottom=False)
ax.tick_params(axis='y', labelsize=11)


ax.grid(axis='x', linestyle='-', color='black')
ax.set_axisbelow(True)
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x:.0f}%'))
ax.set_axisbelow(True)

ax.set_title(column, fontsize=16, color='black', pad=20)
ax.set_xlabel("Percentage of Respondents", fontsize=12, color='black')
ax.set_ylabel("Response", fontsize=12, color='black')

plt.savefig("menstrual_leave.png", transparent=True, bbox_inches='tight')
plt.tight_layout()
plt.show()

files.download("menstrual_leave.png")



```

```python

# Custom color palette
pink_hues = [
    "#FFE4E1",  # Misty Rose
    "#FFB6C1",  # Light Pink
    "#FF69B4",  # Hot Pink
    "#FF1493",  # Deep Pink
    "#FF85A2",  # Soft Rose Pink
    "#FF6EB4",  # Bubblegum Pink
    "#F06292",  # Medium Pink (Material Design)
    "#EC407A",  # Bright Pink
    "#E91E63",  # Raspberry Pink
    "#C2185B",  # Dark Pink / Crimson Pink
]

# Define the correct column
col = "Do you think menstrual leave should be legally mandated in India across all sectors?"

# Desired response order
ordered_labels = ["Yes", "Maybe", "No"]

# Clean and group data
grouped = df[[col, 'Gender']].dropna()

# Unique genders
genders = grouped['Gender'].dropna().unique()

# Create subplots
fig, axs = plt.subplots(1, len(genders), figsize=(8 * len(genders), 6))

# Plot each gender's pie chart
for i, gender in enumerate(genders):
    gender_data = grouped[grouped['Gender'] == gender][col]
    value_counts = gender_data.value_counts()

    # Ensure fixed order
    counts_ordered = [value_counts.get(label, 0) for label in ordered_labels]
    labels = ordered_labels
    colors = pink_hues[:len(labels)]

    axs[i].pie(
        counts_ordered,
        labels=None,
        autopct='%1.1f%%',
        colors=colors,
        startangle=140
    )
    axs[i].legend(labels, title="Responses", loc='center left', bbox_to_anchor=(1, 0.5))
    axs[i].set_title(f"{gender}'s Opinion on Legal Mandate for Menstrual Leave")

plt.savefig("menstrual_leave_pie.png", transparent=True, bbox_inches='tight')
plt.tight_layout()
plt.show()
files.download("menstrual_leave_pie.png")
```

```python
# Column to plot
col = "Have you ever missed school/college/work due to periods?"

# Filter and count responses
response_order = ["Yes", "No", "Maybe"]
value_counts = df[col].dropna().value_counts()
value_counts = value_counts.reindex(response_order).dropna()
print(value_counts)

# Colors
femella_colors = [
    "#DA70D6", "#BA55D3", "#9932CC", "#800080",
    "#8A2BE2", "#4B0082", "#6A0DAD", "#7B1FA2"
]

colors = femella_colors[:len(value_counts)]

# Plotting
fig, ax = plt.subplots(figsize=(6, 6))
wedges, texts, autotexts = ax.pie(
    value_counts,
    labels=None,  # No labels on the slices
    colors=colors,
    autopct='%1.1f%%',
    startangle=140
    )

# Legend instead of labels
ax.legend(wedges, value_counts.index, title="Response", loc="center left", bbox_to_anchor=(1, 0.5), frameon = False)

# Style
plt.setp(autotexts, size=12, weight="bold")
ax.set_title(" ", fontsize=14, color='black')
fig.patch.set_alpha(0.0)  # Transparent background
# Equal aspect ratio ensures pie is a circle
ax.axis('equal')

plt.title("Missed Work/School Due to Periods", fontsize=14, fontweight='bold')

plt.savefig("missed_work_school_periods.png", transparent=True, bbox_inches='tight')
plt.tight_layout()
plt.show()
files.download("missed_work_school_periods.png")
```

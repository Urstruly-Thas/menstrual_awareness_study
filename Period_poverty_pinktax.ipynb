```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mticker
```

```python

# Column to visualize
col = "Do you think access to menstrual hygiene products should be subsidized or made free by the government?"
response_counts = df[col].value_counts()
response_percents = response_counts / response_counts.sum() * 100

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(response_percents.index, response_percents.values, color=pink_hues[1:4])

# Add percentage labels
for bar in bars:
    width = bar.get_width()
    ax.text(width + 1, bar.get_y() + bar.get_height()/2,
            f'{width:.1f}%', va='center', fontsize=11, color="#C2185B",fontweight = "bold")

# Formatting
ax.set_title("Should Menstrual Products Be Subsidized?", fontsize=16, color='black', pad=20)
ax.set_xlabel("Percentage of Respondents", fontsize=12, color='black')
ax.set_ylabel("Response", fontsize=12, color='black')

ax.spines['bottom'].set_color('black')
ax.spines['left'].set_color('black')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.tick_params(colors='black')
ax.xaxis.label.set_color('black')
ax.set_yticklabels(response_percents.index, rotation=360, color='black')
ax.yaxis.label.set_color('black')
ax.title.set_color('black')

ax.grid(axis='x', linestyle='-', color='black')
ax.set_axisbelow(True)
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x:.0f}%'))


# Save with transparent background
plt.tight_layout()
plt.savefig("menstrual_products_subsidy_opinion.png", dpi=300, transparent=True)
plt.show()

files.download("menstrual_products_subsidy_opinion.png")
```

```python
# Drop rows with missing gender or response
df_cleaned = df[['Gender', 'Do you think access to menstrual hygiene products should be subsidized or made free by the government?']].dropna()

# Rename for convenience
df_cleaned.columns = ['Gender', 'SubsidyOpinion']

# Value counts by gender
value_counts = df_cleaned.groupby('Gender')['SubsidyOpinion'].value_counts().unstack().fillna(0).astype(int)

# Percentages by gender
percentages = value_counts.div(value_counts.sum(axis=1), axis=0) * 100

# Display results
print("Counts:")
print(value_counts)
print("Percentages (%):")
print(percentages.round(2))
```

```python

# Teal hues palette
teal_hues = [
    "#E0F7F4", "#B2DFDB", "#80CBC4", "#4DB6AC", "#26A69A",
    "#009688", "#00897B", "#00796B", "#00695C", "#004D40"
]

# Extract the responses and drop NaNs
col = "What challenges do you face during menstruation?"
responses = df[col].dropna()

# Split comma-separated responses and count frequencies
from collections import Counter
all_challenges = []
for response in responses:
    parts = [p.strip() for p in str(response).split(',')]
    all_challenges.extend(parts)

challenge_counts = Counter(all_challenges)
labels = list(challenge_counts.keys())
values = list(challenge_counts.values())
total = sum(values)
percentages = [f"{(v / total * 100):.1f}%" for v in values]
explode= (0.2,0,0,0,0,0)
# Plot pie chart
fig, ax = plt.subplots(figsize=(8, 8), dpi=100)
wedges, texts, autotexts = ax.pie(
    values,
    labels=None,
    colors=teal_hues[:len(labels)],
    startangle=140,
    autopct='%1.1f%%',
    textprops={'color': 'black', 'fontsize': 12}, explode = explode
)

# Add a legend instead of labels
ax.legend(
    wedges,
    labels,
    title="Challenges",
    loc="center left",
    bbox_to_anchor=(1, 0.5),
    fontsize=10,
    title_fontsize=12,frameon = False
)

# Style
plt.setp(autotexts, size=12, weight="bold")
ax.set_title(" ", fontsize=14, color='black')
fig.patch.set_alpha(0.0)  # Transparent background

# Make spines and axis black
ax.spines['bottom'].set_color('black')
ax.spines['left'].set_color('black')

# Save as transparent PNG
plt.savefig("menstrual_challenges_pie.png", transparent=True, bbox_inches='tight')
plt.show()

files.download("menstrual_challenges_pie.png")
```

```python

# Data
labels = ['Access to Menstrual Hygiene Products']
categories = ['Not Always', 'Sometimes']
percentages = [93.83, 6.17]

# Teal shades for stacked segments
teal_hues = [
    "#26A69A",  # For 'Not Always'
    "#80CBC4",  # For 'Sometimes'
]

# Plot setup
fig, ax = plt.subplots(figsize=(8, 1.5))

# Plot the stacked bar
left = 0
for i in range(len(categories)):
    ax.barh(labels, percentages[i], left=left, color=teal_hues[i])
    # Add percentage text in the center of each segment
    ax.text(left + percentages[i]/2, 0,
            f"{percentages[i]:.1f}%",
            ha='center', va='center', fontsize=11,
            color='white', fontweight='bold')
    left += percentages[i]

# Clean aesthetics
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.tick_params(left=False, bottom=False)
ax.set_xticks([])
ax.set_yticks([])
ax.legend(categories, loc='center left', frameon= True )
plt.box(False)
plt.tight_layout()
plt.savefig("Access_to_products.png", transparent=True, bbox_inches='tight')
plt.show()

files.download("Access_to_products.png")
```
